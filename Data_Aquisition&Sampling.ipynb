```python
import pandas as pd
import glob
```

```python
csv_files = glob.glob(r'G:\Mohanraj D_OFFICIAL\GUVI (Data Analyst)\Mini-Projects\PatrolIQ\*.csv')     # This will create a list of path of each csv file present in my PatrolIQ folder 

list_df = []                                                  # An empty list to store all dataframes
for file in csv_files:
    read = pd.read_csv(file)
    list_df.append(read)

df = pd.concat(list_df, ignore_index=True)                    # create one dataframe
```

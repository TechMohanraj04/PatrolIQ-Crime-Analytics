```python
# required datetime conversion
df['Date'] = df['Date'].astype('datetime64[ns]')
df['Updated On'] = df['Updated On'].astype('datetime64[ns]')

# integer type conversion
df['ID'] = df['ID'].astype('Int64')
df['Ward'] = df['Ward'].astype('Int64')
df['District'] = df['District'].astype('Int64')
df['Community Area'] = df['Community Area'].astype('Int64')

# categorical conversion

df["Primary Type"] = df["Primary Type"].astype("category", )
df["Description"] = df["Description"].astype("category")
df["Location Description"] = df["Location Description"].astype("category")
df["FBI Code"] = df["FBI Code"].astype("category")
df["Block"] = df["Block"].astype("category")
df["IUCR"] = df["IUCR"].astype("category")

```

```python
df.isna().sum()
```

```python
# since null values are less than 5% we are dropping rows with null values

df = df.dropna(subset=["X Coordinate"])
df = df.dropna(subset=["Community Area"])
df = df.dropna(subset=["Location Description"])
df = df.dropna(subset=["District"])
df = df.dropna(subset=["Ward"])
```
